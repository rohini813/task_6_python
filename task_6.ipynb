```python
#Smart Parking Lot Management System
```

```python
def parking_system(capacity, logs):
  parked=0
  peak=0

  for log in logs:
    if log=='IN':
      parked+=1
    elif log=='OUT':
      parked-=1
    if parked>peak:
      peak=parked

  print("Currently Parked Vehicles:",parked)

  if parked>capacity:
    print("Parking Status: Full/Over Capacity")
  else:
    print("Parking Status: Available")

capacity=50
vehicle_logs=["IN","IN","OUT","IN","IN","IN","OUT"]
parking_system(capacity,vehicle_logs)
```

```python
#Online Food Delivery Time Estimator
```

```python
def delivery_time(distance, traffic, weather):
  time=distance*5

  if traffic=="High":
    time+=10
  elif traffic=="Medium":
    time+=5

  if weather=="Rainy":
    time+=5

  print("Estimated Time of Delivery:",time,"Minutes")

delivery_time(8,"High","Rainy")
```

```python
#Movie Theatre Seat Occupancy Analyzer
```

```python
def theatre_status(total_seats, booked_list):
  booked=len(booked_list)

  occupancy=(booked/total_seats)*100

  print("Occupancy:",occupancy,"%")

  if occupancy==100:
    print("Show Status : Housefull")
  elif occupancy>=75:
    print("Show Status:Almost full")
  else:
    print("Show Status:Seats Available")

booked_seats=[1]*150
theatre_status(200,booked_seats)
```

```python
#cloud server load classification System
```

```python
def server_load(cpu_readings):
  total=0

  for value in cpu_readings:
    total+=value

  avg=total/len(cpu_readings)

  print("Average CPU Load:",avg,"%")

  if avg <50:
    status="Normal"
  elif avg<=80:
    status="Waring"
  else:
    status="Critical"

  print("Server Status:",status)

cpu=[45, 60, 70, 85, 90]
server_load(cpu)
```

```python
#Smart classroom Resource Usage Monitor
```

```python
def resource_monitor(resources):
    overused=[]

    for item in resources:
        if resources[item]>8:
          overused.append(item)

    print("Overused Resources:",overused)

    if len(overused)>0:
       print("Energy Alert: Yes")
    else:
      print("Energy Alert: NO")

usage={
    "Projector":6,
    "AC":9,
    "Lights":4
}

resource_monitor(usage)
```

```python
#Online Event Registration Capacity Controller
```

```python
def event_registration(capacity,events):
   if events<=capacity:
      confirmed=events
      waitlist=0
      status="Open"
   else:
      confirmed=capacity
      waitlist=events-capacity
      status="Closed"

   print("Confirmed Registrations:",confirmed)
   print("Waitlist Registrations:",waitlist)
   print("Event Status:",status)

event_registration(100,105)
```
